```python
from process import *
from plots import *
from envelopes import *
from supports import *
from plotsettings import *

import glob, os
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Data load - BS interaction
file_BS_full = glob.glob("*200715_195616*")[0]
file_BS_rwa = glob.glob("*200803_062445*")[0]
file_BS_cr = glob.glob("*200716_163328*")[0]
```

```python
# Data plot - BS interaction

BS_full =  file_BS_full
BS_rwa =  file_BS_rwa
BS_cr =  file_BS_cr

quants = ['times', 'expect', 'e0', 'g1', 'e1', 'g0']

BS_full_times, _, BS_full_expect, BS_full_e0, BS_full_g1, _  , _      , _ = load_data(quants, file_BS_full)
BS_rwa_times, _, BS_rwa_expect, BS_rwa_e0, BS_rwa_g1, _  , _      , _ = load_data(quants, file_BS_rwa)
BS_cr_times, _, BS_cr_expect, BS_cr_e0, BS_cr_g1, _  , _      , _ = load_data(quants, file_BS_cr)

BS_full_times = BS_full_times-20
BS_rwa_times = BS_rwa_times[0:len(BS_full_times )]-20
BS_cr_times = BS_cr_times[0:len(BS_full_times )]-20

BS_rwa_e0 = BS_rwa_e0[0:len(BS_full_times )]
BS_rwa_g1 = BS_rwa_g1[0:len(BS_full_times )]
BS_cr_e0 = BS_cr_e0[0:len(BS_full_times )]
BS_cr_g1 = BS_cr_g1[0:len(BS_full_times )]

fig = plt.figure(figsize=[22,4])
plt.plot(BS_full_times, BS_full_e0-BS_full_g1, c='r', alpha=alpha, label="Full")
plt.plot(BS_rwa_times, BS_rwa_e0-BS_rwa_g1, c='r',ls='dashed', alpha=alpha, label="RWA")
plt.plot(BS_cr_times, BS_cr_e0-BS_cr_g1, c='r',ls='dashdot', alpha=alpha, label="CR")

plt.xlim([0, 400])
plt.ylim([-1, 1.0])
plt.yticks([-1,-0.5,0,0.5,1],fontsize=20)
plt.xticks(fontsize=25)
plt.text(15,-0.5,"(a)",fontsize=30)
plt.legend(loc='lower right',bbox_to_anchor=(1, 0),fontsize=22)
plt.xlabel("Pulse width (ns)",fontsize=25)
plt.ylabel("$P(e0)-P(g1)$",fontsize=20)

plt.show()
```
