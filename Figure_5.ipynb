```python
# Data load.

quants = ['times', 'e1', 'g0', 'coupling', 'expect']

folder1 = "data_200511_074439_blue_Nq10_Nt1_eps0.6_wd5.216_wq6.5_wc4.0_wsb0.0"  # Data for figure part (a) and (b)
srcfolder1 = home + folder1
times1, _,_,_,_, e1_1, g0_1, coupling1 = load_data(quants, srcfolder1)

folder2 = "data_191108_084954_blue_Nq10_Nt1_Om0.6_wd5.216_wq6.5_wc4.0_wsb0.0037" # Data for figure part (c) and (d)
srcfolder2 = home + folder2
times2,  _,_,_,_, e1_2, g0_2, coupling2 = load_data(quants, srcfolder2)
```
