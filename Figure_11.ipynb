```python
import numpy as np
import matplotlib.pyplot as plt
import pickle
```

# Data Load and Plotting

```python
# Data load.
load_file1 = open('UnShift.dat', 'rb')    # resonator transmission without transmon saturation.
loaddata1 = pickle.load(load_file1)

load_file2 = open('Shift.dat', 'rb')      # resonator transmission with transmon saturation.
loaddata2 = pickle.load(load_file2)


## Plot.
plt.figure()
plt.plot(loaddata1['probe_freq'], loaddata1['transmission_data'], 'k-')
plt.plot(loaddata2['probe_freq'], loaddata2['transmission_data'], 'k--')
plt.xlim([4.07,4.08])
plt.ylim(0.0008,0.0014)
plt.yticks([0.0008,0.0011,0.0014],fontsize=12)
plt.xticks([4.07,4.0725,4.075,4.0775,4.08],fontsize=15)
plt.xlabel(r'$\omega_{p}~$($2 \pi\cdot$GHz)',fontsize=15)
plt.ylabel(r'S$_{21}$dB (arb. units)',fontsize=15)
plt.tight_layout()
plt.savefig('Dispersive_shift.pdf',dip=300)
plt.show()
```
